Simplificamos los valores categoricos

```python
def _convertir_titulo(X):
    titles = {
        "Mr": ["Mr", "Master", "Dr", "Col", "Capt", "Jonkheer", "Don", "Rev", "Sir", "Major"],
        "Mrs": ["Mrs", "Ms", "Mme", "Lady", "Mlle", "Countess"],
        "Miss": ["Miss"],
    }

    for key, values in titles.items():
        if X in values:
            return key

    return X


df["Title"] = df["Title"].apply(_convertir_titulo)
df["Title"].unique().tolist()
```

Ahora procedemos a convertir las columnas booleanas

```python
potential_boolean_values = [
    [0, 1],
    [0.0, 1.0],
    ["false", "true"],
    ["no", "yes"],
    ["no", "si"],
    ["N", "Y"],
]

boolean_sets = [set(map(str, values_list)) for values_list in potential_boolean_values]

def is_boolean_column(unique_values):
    unique_values = set(map(str, unique_values))
    return any(unique_values == boolean_set for boolean_set in boolean_sets)

def convert_to_boolean(df, column):
    false_values = set(map(str, [0, 0.0, "false", "no", "N"]))
    df[column] = df[column].apply(lambda x: x if pd.isna(x) else str(x).lower() in false_values)

columns = df.columns.tolist()
converted_boolean_columns = []

for column in columns:
    unique_values = df[column].dropna().unique()
    if len(unique_values) == 2 and is_boolean_column(unique_values):
        print(f"Column '{column}' es booleana, lo convertimos a tipo booleana.")
        convert_to_boolean(df, column)
        df[column] = df[column].astype("bool")
        converted_boolean_columns.append(column)

print(f"Columna convertida a booleana: {converted_boolean_columns}")
```

```python
potential_categorical_columns = df.select_dtypes(include=["object", "string"]).columns.tolist()

potential_categorical_columns
```

Eliminamos espacios y caracteres invalidos de las cadenas

```python
from utils.eda import clean_string


for column in potential_categorical_columns:
    df[column] = df[column].apply(clean_string)
```

Convertimos las columnas a categorica o cadena de acuerdo al porcentaje de valores unicos en comparación al total

```python
max_percentage_categorical = 0.05
num_rows_df = df.shape[0]
converted_to_categorical = []
converted_to_string = []
for column in potential_categorical_columns:
    unique_values_count = df[column].nunique(dropna=True)
    print(f"Cantidad de valores unicos en la columna '{column}': {unique_values_count}")
    if unique_values_count <= (max_percentage_categorical * num_rows_df):
        df[column] = df[column].astype("category")
        converted_to_categorical.append(column)
    else:
        df[column] = df[column].astype("string")
        converted_to_string.append(column)
print("\n")
print(f"Número de columnas analizadas: {len(potential_categorical_columns)}")
print(f"Número de columnas convertidas a category: {len(converted_to_categorical)}")
print(f"Número de columnas convertidas a string: {len(converted_to_string)}")
print(f"Columnas convertidas a category: {converted_to_categorical}")
print(f"Columnas convertidas a string: {converted_to_string}")

```

Eliminamos las columnas que no utilizaremos y que no aportan información importante

```python
df = df.drop("PassengerId", axis=1)
df = df.drop("Name", axis=1)
df = df.drop("Ticket", axis=1)
```

Creamos la característica FamilySize

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1 # el 1 incluye al pasajero
```

```python
def _convert_unknown_cabin(X):
    if X == "unk":
        return True

    return False


df["CabinUnknown"] = df["Cabin"].apply(_convert_unknown_cabin)
```

```python
get_descriptive_statistics(df).T    
```

```python
def _convert_fare(X):
    if X < 50:
        return "basic"
    
    elif 50 <= X <= 250:
        return "medium"

    else:
        return "premium"


df["FareCat"] = df["Fare"].apply(_convert_fare).astype("category")
df["FareCat"]
```

Creamos una columna para saber el nivel en que se encuentra cada pasajero, en caso de null o dato inconsistente completamos con M

```python
df["Deck"] = df["Cabin"].apply(lambda s: s[0] if pd.notnull(s) else 'M')
```

```python
new_columns = ["Title", "FamilySize", "FareCat", "CabinUnknown", "Deck"]
```

Ahora evaluaremos las características agregadas

```python
columns = df.columns.tolist()
columns
```

```python
original_columns = [
    'Pclass',
    'Sex',
    'Age',
    'SibSp',
    'Parch',
    'Fare',
    'Embarked',
]

feature_columns = [
    'Pclass',
    'Sex',
    'Age',
    'Fare',
    'Embarked',
    'Title',
    'FamilySize',
    'CabinUnknown',
    'FareCat',
    'Deck',
]

columnas_total = list(set(original_columns + feature_columns))

X = df[columnas_total]

y = df["Survived"]
```

```python
from utils.eda import get_categoric_columns


categoric_columns = get_categoric_columns(df)

for column in columnas_total:
    if column not in categoric_columns:
        continue

    unique_values = df[column].unique().tolist()
    print(f"Valores únicos para columna '{column}': {str(unique_values)}")
```

De estos datos, podemos obtener columnas a codificar como ordinales como one hot.

```python
ordinal_features_columns = [
    "FareCat",
]

ordinal_columns_category_feature = [["basic", "medium", "premium"]]

onehot_original_columns = [
    "Sex",
    "Embarked"
]

onehot_feature_columns= [
    "Sex",
    "Title",
    "Deck",
    "Embarked"
]
```

Importamos las librerias para modelar

```python
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, MinMaxScaler
from sklearn.pipeline import Pipeline
import xgboost as xgb
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.model_selection import cross_val_score
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
)
print(X_train.columns)

X_train_original = X_train[original_columns]
X_train_feature = X_train[feature_columns]
X_test_original = X_test[original_columns]
X_test_feature = X_test[feature_columns]
```

Seteamos preprocesadores

```python
print("Columnas en X_train_original:", X_train_original.columns)

# Preprocesador original
preprocessor_original = ColumnTransformer(
    transformers=[
        ("onehot", OneHotEncoder(), onehot_original_columns)
    ],
    remainder="passthrough",
    force_int_remainder_cols=False
)

# Preprocesador con Feature Engineering
preprocessor_feature_eng = ColumnTransformer(
    transformers=[
        ("ordinal", OrdinalEncoder(), ordinal_features_columns),
        ("onehot", OneHotEncoder(), onehot_feature_columns),
    ],
    remainder="passthrough",
    force_int_remainder_cols=False
)

```

#### Modelo sin Feature engineering

```python
rf_original = Pipeline(
    steps=[
        ("preprocessor", preprocessor_original),
        ("model", RandomForestClassifier(n_estimators=20, random_state=42))
    ]
)

# Ajustar el modelo

rf_original.fit(X_train_original, y_train)
```

Evaluamos resultados

```python
y_pred_rf_original = rf_original.predict(X_test_original)

accuracy_rf_original = accuracy_score(y_test, y_pred_rf_original)
precision_rf_original = precision_score(y_test, y_pred_rf_original)
recall_rf_original = recall_score(y_test, y_pred_rf_original)
f1_rf_original = f1_score(y_test, y_pred_rf_original)
roc_auc_rf_original = roc_auc_score(y_test, y_pred_rf_original)

print(f"Accuracy RF original: {accuracy_rf_original}")
print(f"Precision RF original: {precision_rf_original}")
print(f"Recall RF original: {recall_rf_original}")
print(f"F1 RF original: {f1_rf_original}")
print(f"ROC AUC original: {roc_auc_rf_original}")
```

#### Modelo con Feature Engineering

```python
rf_feature_eng = Pipeline(
    steps=[
        ("preprocessor", preprocessor_feature_eng),
        ("model", RandomForestClassifier(n_estimators=20, random_state=42))
    ]
)

rf_feature_eng.fit(X_train_feature, y_train)
```

Evaluamos resultados

```python
y_pred_rf_feature_eng = rf_feature_eng.predict(X_test_feature)

accuracy_rf_feature_eng = accuracy_score(y_test, y_pred_rf_feature_eng)
precision_rf_feature_eng = precision_score(y_test, y_pred_rf_feature_eng)
recall_rf_feature_eng = recall_score(y_test, y_pred_rf_feature_eng)
f1_rf_feature_eng = f1_score(y_test, y_pred_rf_feature_eng)
roc_auc_rf_feature_eng = roc_auc_score(y_test, y_pred_rf_feature_eng)


print(f"Accuracy RF Feature Eng: {accuracy_rf_feature_eng}")
print(f"Precision RF Feature Eng: {precision_rf_feature_eng}")
print(f"Recall RF Feature Eng: {recall_rf_feature_eng}")
print(f"F1 RF Feature Eng: {f1_rf_feature_eng}")
print(f"ROC AUC Feature Eng: {roc_auc_rf_feature_eng}")
```

Comparamos el rendimiento de ambos modelos

```python
rf_original_metrics = {
    "Accuracy": accuracy_rf_original,
    "Precision": precision_rf_original,
    "Recall": recall_rf_original,
    "F1-Score": f1_rf_original,
    "ROC-AUC": roc_auc_rf_original,
}

rf_feature_eng_metrics = {
    "Accuracy": accuracy_rf_feature_eng,
    "Precision": precision_rf_feature_eng,
    "Recall": recall_rf_feature_eng,
    "F1-Score": f1_rf_feature_eng,
    "ROC-AUC": roc_auc_rf_feature_eng,
}

results = pd.DataFrame(
    [rf_original_metrics, rf_feature_eng_metrics,],
    index=["Random Forests Original", "Random Forests Feature",],)
print(results)

print("\n")

print("Mejores modelos por métrica:\n")
print(results.idxmax(axis=0))
```

Aplicando Feature engineering se obtienen mejores métricas
